```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_x = pd.read_csv('../../3_MeV_Positron_XAxis_nt_PhotonData.csv', skiprows=7, header=None, names=['Count', 'x'])
df_x.head(10)
```

```python
# Группировка данных по x и вычисление среднего и стандартной ошибки среднего
grouped_x = df_x.groupby('x')['Count'].agg(['mean', 'std', 'count']).reset_index()
grouped_x['error'] = grouped_x['std'] / np.sqrt(grouped_x['count'])  # Стандартная ошибка среднего
```

```python
# Построение графика
plt.errorbar(grouped_x['x'], grouped_x['mean'], yerr=grouped_x['error'], fmt='o', capsize=5, label='Mean with error')
plt.xlabel('X, mm')
plt.ylabel('Count')
plt.title('Detector response function in X_axis mode [31 points with 1000 events in each point]')
plt.grid(True)
plt.show()
```
